# Data Augmentation

- **Handling information loss due to stride**:
  - There might be some important information that we unintentionally removed by using stride. This can be tackled using data augmentation, where we create multiple images from one, so somehow the information gets captured.
  - In the above example, the last column is completely ignored, which might result in loss of information.
  - If the stride is large, we might lose important information.

- **Using data augmentation, we can achieve invariance**:
  - Ensures that the model is less sensitive to small changes in the input data, enhancing robustness.

- **Invariance concept**:
  - The output remains consistent even if the input changes slightly, improving the model's reliability and performance.
  
- **If we have a small dataset, we can easily make it larger**:
  - Generates additional training samples, effectively expanding the dataset size without collecting new data.

- **It also solves the overfitting problem**:
  - Reduces overfitting by introducing variability, ensuring the model generalizes better to unseen data.

- **CNN model robustness**:
  - Data augmentation techniques such as rotation, scaling, zooming in/out, horizontal flip, and vertical flip enhance the model's ability to handle various transformations.



### Summary
Data augmentation is a powerful technique to enhance model performance by increasing dataset size, reducing overfitting, and achieving invariance, making CNN models robust to various transformations and small input changes. Additionally, it helps in recovering important information that might be lost due to the use of stride in convolutional layers.

```python
# Data Augmentation package/library
# from keras.preprocessing.image import ImageDataGenerator
import keras
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from keras.utils import img_to_array, array_to_img, load_img
import matplotlib.pyplot as plt
```

```python
# importing the image
image_file = '/Users/nidhisingh/iCloud Drive (Archive)/Documents/sample_datasets/cool_cat.jpeg'
dog_image = keras.preprocessing.image.load_img(image_file)
dog_image
```

```python
dog_array = keras.preprocessing.image.img_to_array(dog_image)
dog_array
```

```python
# Defining original image vs Transformed image

def plot_image(original_image, updated_image):
    fig, axes = plt.subplots(1, 2, figsize=(10, 5))

    # original image
    axes[0].set_title("Original Image")
    axes[0].imshow(original_image)

    # Transformed image
    axes[1].set_title("Transformed Image")
    axes[1].imshow(updated_image)

    plt.tight_layout()
    plt.show()
```

```python
# using single transform feature - Flip an image 
flipped = ImageDataGenerator(horizontal_flip=False, vertical_flip=True)
transformed_array = flipped.random_transform(dog_array)
transformed_image =  array_to_img(transformed_array)
plot_image(dog_image, transformed_image)
```

```python
zoomed = ImageDataGenerator(zoom_range=1.0)
transformed_array = zoomed.random_transform(dog_array)
transformed_image = array_to_img(transformed_array)
plot_image(dog_image, transformed_image)
```

```python
height_shift = ImageDataGenerator(height_shift_range=155)
transformed_array = height_shift.random_transform(dog_array)
transformed_image = array_to_img(transformed_array)
plot_image(dog_image, transformed_image)
```

```python
# Applying multiple transforms
combined_transformation = ImageDataGenerator(height_shift_range=0.2,width_shift_range=0.2,
                                                            zoom_range=0.3,horizontal_flip=True,vertical_flip=True,
                                                          rotation_range=30, shear_range=0.2  )
transformed_array = combined_transformation .random_transform(dog_array)
transformed_image = array_to_img(transformed_array)
plot_image(dog_image, transformed_image)
```
